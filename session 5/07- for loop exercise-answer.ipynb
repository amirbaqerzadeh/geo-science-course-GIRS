```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x-axis values
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

# Define the mathematical functions
functions = [
    np.sin,  # y = sin(x)
    np.cos,  # y = cos(x)
    np.tan,  # y = tan(x)
    lambda x: x**2,  # y = x^2
]

# Titles for each subplot
titles = [
    "y = sin(x)",
    "y = cos(x)",
    "y = tan(x)",
    "y = x^2",
]

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Mathematical Functions", fontsize=16)

# Loop through the functions and plot them
for i, (ax, func, title) in enumerate(zip(axes.flat, functions, titles)):
    y = func(x)  # Compute y values
    ax.plot(x, y, label=title, color=f"C{i}")  
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()
    ax.grid(True)

    
    if func == np.tan:
        ax.set_ylim(-10, 10)  # Limit y-axis for tan(x)


plt.tight_layout()

# Save the figure
fig.savefig("multi_plot.png", bbox_inches="tight")

# Show the plot
plt.show()
```
