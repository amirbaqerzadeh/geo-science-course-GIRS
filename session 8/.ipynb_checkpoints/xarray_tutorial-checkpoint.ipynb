```python
file_path = './data/icec.day.mean.1981.nc'
ds = xr.open_dataset(file_path)
```

```python
ds
```

```python
print(ds)
```

```python
icec = ds['icec']
```

```python
type(icec)
```

```python
print(icec.dims)
```

```python
print(icec.attrs)
```

```python
type(icec.attrs)
```

```python
icec.isel(time=0)
```

```python
icec.isel(time=0).plot()
```

```python
global_mean = icec.mean(dim=['lat', 'lon'])
```

```python
global_mean.plot()
```

```python
icec.plot(col='time', col_wrap=6, cmap='Blues')
plt.show()
```

```python
specific_date = icec.sel(time='1981-09-15')
specific_date.plot()

plt.show()
```

```python
region = icec.sel(lat=slice(60, 90), lon=slice(0, 360))
region.isel(time=0).plot()
```

```python
# Time series of Arctic sea ice concentration
arctic_mean = icec.mean(dim=['lat', 'lon'])
arctic_mean.plot(marker='o')
plt.title("Arctic Mean Sea Ice Concentration (1981)")
plt.xlabel("Time")
plt.ylabel("Ice Concentration (%)")
plt.show()
```

```python
save_path = "processed_data.nc"
arctic_mean.to_netcdf(save_path)
```
